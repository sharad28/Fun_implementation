```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
car_path = "/content/drive/MyDrive/Colab Notebooks/Deep learning/Day 14/car1.jpeg"
```

```python
car = cv2.imread(car_path)
car_rgb = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(car_rgb,cmap="gray")
```

```python
car_rgb.shape
```

```python
conv_layer = tf.keras.layers.Conv2D(filters=1,kernel_size=3,input_shape=car_rgb.shape)
model = tf.keras.Sequential(conv_layer)
model.summary()
```

```python
row,col,depth = car_rgb.shape
print(car_rgb.shape)
car_rgb_reshape = car_rgb.reshape(1,row,col,depth) 
```

```python
out = model.predict(car_rgb_reshape)
_,row,col,depth = out.shape
```

```python
out= out.reshape(row,col,depth)
# plt.imshow(out)
plt.imshow(out[:,:,0],cmap="gray")
```
